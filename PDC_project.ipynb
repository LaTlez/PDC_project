# Define Orthonormal Signal Waveform by the picture 1

```python
s1(t) = 2 * log(x2) * (heaviside(t) - heaviside(t - (log(2 * x3) / 4)))


print("Value in x-axis: ", RR(log(2 * x3) / 4).n(digits = 2))
print("Value in y-axis: ", RR(2 * log(x2)).n(digits = 2))

plot(s1(t), t, -1, 3, figsize = (3, 2))
```

```python
s2(t) = 2 * log(x2) * (heaviside(t - (log(2 * x3) / 4)) - heaviside(t - (log(2 * x3) / 2)))


print("Value in x1-axis: ", RR(log(2 * x3) / 4).n(digits = 2))
print("Value in x2-axis: ", RR(log(2 * x3) / 2).n(digits = 2))
print("Value in y-axis: ", RR(2 * log(x2)).n(digits = 2))

plot(s2(t), t, -1, 3, figsize = (3, 2))
```

```python
s3(t) = 2 * log(x2) * (heaviside(t - (log(2 * x3) / 2)) - heaviside(t - (3 * log(2 * x3) / 4)))


print("Value in x1-axis: ", RR(log(2 * x3) / 2).n(digits = 2))
print("Value in x2-axis: ", RR(3 * log(2 * x3) / 4).n(digits = 2))
print("Value in y-axis: ", RR(2 * log(x2)).n(digits = 2))

plot(s3(t), t, -1, 3, figsize = (3, 2))
```

# Define Basic Function Signal Waveform by the picture 2

```python
ψ1(t) = 2 / sqrt(log(2 * x3)) * (heaviside(t) - heaviside(t - (log(2 * x3) / 4)))

print("Value in x-axis: ", RR(log(2 * x3) / 4).n(digits = 2))
print("Value in y-axis: ", RR(2 / sqrt(log(2 * x3))).n(digits = 2))

plot(ψ1(t), t, -1, 3, figsize = (3, 2))
```

```python
ψ2(t) = 2 / sqrt(log(2 * x3)) * (heaviside(t - (log(2 * x3) / 4)) - heaviside(t - (log(2 * x3) / 2)))


print("Value in x1-axis: ", RR(log(2 * x3) / 4).n(digits = 2))
print("Value in x2-axis: ", RR(log(2 * x3) / 2).n(digits = 2))
print("Value in y-axis: ", RR(2 / sqrt(log(2 * x3))).n(digits = 2))

plot(ψ2(t), t, -1, 3, figsize = (3, 2))
```

```python
ψ3(t) = 2 / sqrt(log(2 * x3)) * (heaviside(t - (log(2 * x3) / 2)) - heaviside(t - (3 * log(2 * x3) / 4)))


print("Value in x1-axis: ", RR(log(2 * x3) / 2).n(digits = 2))
print("Value in x2-axis: ", RR(3 * log(2 * x3) / 2).n(digits = 2))
print("Value in y-axis: ", RR(2 / sqrt(log(2 * x3))).n(digits = 2))

plot(ψ3(t), t, -1, 3, figsize = (3, 2))
```
